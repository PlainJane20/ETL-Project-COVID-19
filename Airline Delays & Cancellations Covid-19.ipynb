```python
number_of_delayed = q1flight20_data["dep_delay"].apply(lambda s: 1 if s!=0 else 0);
print("Total number of flights: "+str(len(q1flight20_data)))
print("Number of cancelled flights: "+str(sum(q1flight20_data["cancelled"])))
print("Number of delayed flights: "+str(sum(number_of_delayed)))
print("Number of diverted flights: "+str(sum(q1flight20_data["diverted"])))

print("Number of not cancelled flights: "+str(len(q1flight20_data)-sum(q1flight20_data["cancelled"])))
print("Number of not delayed flights: "+str(len(q1flight20_data)-sum(number_of_delayed)))

# print("The number of missing data: "+str(q1flight20_data['actl_deptime'].isnull().sum()));

print("Percentage of cancelled flights: "+str((sum(q1flight20_data["cancelled"])*1.0/len(q1flight20_data))*100)+"%")
print("Percentage of delayed flights: "+str((sum(number_of_delayed)*1.0/len(q1flight20_data))*100)+"%")
```

```python
q1flight20_data["on_time"]=q1flight20_data["arr_delay"].apply(lambda row: 1 if row==0 else 0)
print(len(q1flight20_data["carrier_delay"]))
print("on_time: "+str(q1flight20_data["on_time"].sum()))
missing_data_info={};
for column in q1flight20_data.columns:
    missing_data_info[column]=q1flight20_data[column].isnull().sum()
missing_data_info_sorted = sorted(missing_data_info.items(), key=operator.itemgetter(1))
missing_data_info_sorted
```

```python
cancelled_flights = q1flight20_data
grouped_cancelled_flights=cancelled_flights[["airline","cancelled","on_time"]].groupby(['airline']).sum().reset_index()
grouped_cancelled_flights["flights_count"]=cancelled_flights[["airline","on_time"]].groupby(['airline']).count().reset_index()["on_time"]
grouped_cancelled_flights["cancelled_pct"]=grouped_cancelled_flights["cancelled"]*1.0/grouped_cancelled_flights["flights_count"]*100
grouped_cancelled_flights["on_time_pct"]=grouped_cancelled_flights["on_time"]*1.0/grouped_cancelled_flights["flights_count"]*100
grouped_cancelled_flights[["airline","flights_count","cancelled","on_time","cancelled_pct","on_time_pct"]].sort_values(by=['cancelled_pct'],ascending=[False])
```

* Added new column flights_count to usairlines data which represents the total count of flights
* Added new column on_time to usairlines data that represents the count of flights that were on time
* Added new column on_time_pct to usairlines data which can be useed to sort the data + decide which airline is better than another

```python
def get_airline_information(column_name,allairlines_data,q1flight20_data):
    return allairlines_data.apply(lambda x: q1flight20_data.loc[x["iata_code"]==q1flight20_data["airline"],column_name].values[0] if len(q1flight20_data.loc[x["iata_code"]==q1flight20_data["airline"],column_name])>0 else 0,axis=1)
```

```python
allairlines_data["flights_count"]=get_airline_information("flights_count",allairlines_data,grouped_cancelled_flights)
allairlines_data["on_time"]=get_airline_information("on_time",allairlines_data,grouped_cancelled_flights)
allairlines_data["on_time_pct"]=get_airline_information("on_time_pct",allairlines_data, grouped_cancelled_flights)
allairlines_data.sort_values(by="on_time_pct",ascending=False)
```

```python
flight_cancelled_count = pd.read_sql_query('SELECT a.airline as `airline name`, a.iata_code as `airline`,\
                                        SUM(f.cancelled) as `total_cancelled`\
                                        FROM usairlines as a\
                                        INNER JOIN Q1mergedfile as f ON a.iata_code = f.airline\
                                        WHERE cancel_reason = "A"\
                                        GROUP BY f.airline\
                                        ORDER BY(total_cancelled) desc;', con = engine)
flight_cancelled_count
```

```python
flight_cancelled_count.to_sql(name='flight_cancelled_count', con=engine, if_exists='replace', index=False)
engine.table_names()
```

```python
cancelled_count = pd.read_sql_query('SELECT usairlines.airline, usairlines.iata_code, Q1mergedfile.cancelled as total_cancelled\
                                     FROM usairlines \
                                     JOIN Q1mergedfile\
                                     ON usairlines.iata_code = Q1mergedfile.airline\
                                     WHERE Q1mergedfile.cancelled = 1;', con = engine)
cancelled_count.to_sql(name='cancelled_count', con=engine, if_exists='replace', index=False)
engine.table_names()
```

```python
cancelled_count.head()
```

```python
total_flight_cancelled = pd.read_sql_query('SELECT airline, iata_code, SUM(total_cancelled) as cancelled_flights\
                                     FROM cancelled_count \
                                     GROUP BY airline\
                                     ORDER BY SUM(total_cancelled) desc;', con = engine)
total_flight_cancelled
```

```python
total_flight_cancelled.to_sql(name='total_cancelled', con=engine, if_exists='replace', index=False)
engine.execute("DROP TABLE cancelled_count")
```

```python
engine.table_names()
```

```python
total_flights = pd.read_sql_query("SELECT airline, COUNT(airline) as flight_totals\
                                     FROM Q1mergedfile \
                                     GROUP BY airline", con = engine)
total_flights.to_sql(name='total_flights', con=engine, if_exists='replace', index=False)
```

```python
engine.table_names()
```

```python
total_flights
```

```python
flightcancelled_count = pd.read_sql_query("SELECT total_cancelled.airline, total_cancelled.iata_code,\
                                        total_cancelled.cancelled_flights,total_flights.flight_totals \
                                        FROM total_cancelled JOIN total_flights \
                                        ON total_cancelled.iata_code = total_flights.airline", con = engine)
flightcancelled_count
```

```python
flightcancelled_count['cancellation ratio'] = round(flightcancelled_count['cancelled_flights'] / flightcancelled_count['flight_totals'] *100,2)
flightcancelled_count = flightcancelled_count.sort_values(['cancellation ratio'])
flightcancelled_count
```

# Airline Delay Time

```python
avg_airline_delays = pd.read_sql_query('SELECT a.airline as `airline Name`, a.iata_code as `airline Code`,\
                  COUNT(f.airline) as `Total Flights not Cancelled`,\
                  AVG(f.dep_delay) as `Avg dep_delay`,\
                  AVG(f.arr_delay) as `Avg arr_delay`\
                  FROM usairlines as a\
                  INNER JOIN Q1mergedfile as f ON a.iata_code = f.airline\
                  WHERE f.cancelled = 0\
                  GROUP BY f.airline\
                  ORDER BY(`AVG dep_delay`) desc;', con = engine)
```

```python
avg_airline_delays
```

```python
avg_airline_delays.to_sql(name='avg_airline_delays', con=engine, if_exists='replace', index=False)
engine.table_names()
```

```python
airline_departure_delay_counts = pd.read_sql_query('SELECT a.airline as `airline_name`, a.iata_code as `airline`,\
                       COUNT(f.dep_delay) as `count_dept_delay`\
                       FROM usairlines as a\
                       INNER JOIN Q1mergedfile as f ON a.iata_code = f.airline\
                       WHERE f.dep_delay <> 0\
                       GROUP BY f.airline\
                       ORDER BY(`count_dept_delay`) desc;', con = engine)
```

```python
airline_departure_delay_counts
```

```python
airline_departure_delay_counts.to_sql(name='airline_departure_delay_counts', con=engine, if_exists='replace', index=False)
engine.table_names()
```

```python
airline_departure_delay_ratio = pd.read_sql_query("SELECT airline_departure_delay_counts.airline_name, airline_departure_delay_counts.airline,\
                                        airline_departure_delay_counts.count_dept_delay, total_flights.flight_totals\
                                        FROM airline_departure_delay_counts JOIN total_flights \
                                        ON airline_departure_delay_counts.airline = total_flights.airline", con = engine)
airline_departure_delay_ratio
```

```python
airline_departure_delay_ratio['departure_delay_ratio'] = round(airline_departure_delay_ratio['count_dept_delay'] / airline_departure_delay_ratio['flight_totals'] *100,2)
airline_departure_delay_ratio = airline_departure_delay_ratio.sort_values(['departure_delay_ratio'])
airline_departure_delay_ratio
```

```python
airline_departure_delay_ratio.to_sql(name='airline_departure_delay_ratio', con=engine, if_exists='replace', index=False)
engine.table_names()
```

```python
airline_arrival_delay_counts = pd.read_sql_query('SELECT a.airline as `airline_name`, a.iata_code as `airline`,COUNT(f.arr_delay) as `count_arrival_delay`\
                    FROM usairlines as a\
                    INNER JOIN Q1mergedfile as f ON a.iata_code = f.airline\
                    WHERE f.arr_delay <> 0\
                    GROUP BY f.airline\
                    ORDER BY(`count_arrival_delay`) desc;', con = engine)
airline_arrival_delay_counts
```

```python
airline_arrival_delay_counts.to_sql(name='airline_arrival_delay_counts', con=engine, if_exists='replace', index=False)
engine.table_names()
```

```python
airline_arrival_delay_ratio = pd.read_sql_query("SELECT airline_arrival_delay_counts.airline_name, airline_arrival_delay_counts.airline,\
                                        airline_arrival_delay_counts.count_arrival_delay, total_flights.flight_totals\
                                        FROM airline_arrival_delay_counts JOIN total_flights \
                                        ON airline_arrival_delay_counts.airline = total_flights.airline", con = engine)
airline_arrival_delay_ratio
```

```python
airline_arrival_delay_ratio['arrival_delay_ratio'] = round(airline_arrival_delay_ratio['count_arrival_delay'] / airline_arrival_delay_ratio['flight_totals'] *100,2)
airline_arrival_delay_ratio = airline_arrival_delay_ratio.sort_values(['arrival_delay_ratio'])
airline_arrival_delay_ratio
```
